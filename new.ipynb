# NASDAQ - AT&T, T-Mobile, Verizon

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
att = pd.read_csv("T.csv")
tm = pd.read_csv("TMUS.csv")
vz = pd.read_csv("VZ.csv")
```

```python
att["Date"] = pd.to_datetime(att['Date'])
tm["Date"] = pd.to_datetime(tm['Date'])
vz["Date"] = pd.to_datetime(vz['Date'])
```

## Comparative Analyses

### Volume

```python
plt.plot(att["Date"], att["Volume"])
plt.plot(tm["Date"], tm["Volume"])
plt.plot(vz["Date"], vz["Volume"])
plt.title("AT&T Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

### Price

```python
plt.plot(att["Date"], att["Close"])
plt.plot(tm["Date"], tm["Close"])
plt.plot(vz["Date"], vz["Close"])
plt.title("AT&T Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

## AT&T

### Volume

```python
plt.plot(att["Date"], att["Volume"])
plt.title("AT&T Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

Notice the spike in volume around the year 1988 in the plot above. 

### Price

```python
plt.plot(att["Date"], att["Close"])
plt.title("AT&T Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

### 1988: An Instance of Volume Movement Preceding Price Movement

```python
att_1988 = att[(att["Date"].dt.year >= 1988) & (att["Date"].dt.year <= 1995)]
```

```python
plt.plot(att_1988["Date"], att_1988['Volume'])
plt.title("AT&T Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```

```python
plt.plot(att_1988["Date"], att_1988['Close'])
plt.title("AT&T Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```
